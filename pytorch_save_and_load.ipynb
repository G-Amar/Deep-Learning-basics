```python
import torch
import torchvision.models as models
```

```python
#pytorch models store the parameters in `state_dict`
#can use torch.save to save these weights
model = models.vgg16(weights='IMAGENET1K_V1')
torch.save(model.state_dict(), 'model_weights.pth')
```

```python
#can load model weights into an untrained model of the same instance using load_state_dict()

model = models.vgg16() # we do not specify ``weights``, i.e. create untrained model
model.load_state_dict(torch.load('model_weights.pth'))
model.eval()
```

```python
#when loading, need model class
#can save model class structure with the weights by passing the entire model (not just `model.state.dict()`)
torch.save(model, 'model.pth')

#then we can load with this, much simpler!
model = torch.load('model.pth')
```
